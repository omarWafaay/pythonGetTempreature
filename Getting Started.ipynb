# Importing necessary packages

```python
# Import Meteostat library and dependencies
from datetime import datetime,timedelta
import matplotlib.pyplot as plt
from meteostat import Point, Daily,Stations
```

```python
# Set time period
"""
#You can choose relative dates from your current time, or choose specific dates
start = datetime(2018, 1, 1)
end = datetime(2018, 12, 31)
"""
start=datetime.now()
end=datetime.now()+timedelta(11)

```

```python
#Chhosing the location
"""
You can add the required location using it's geographical coordinates using the Point() class
"""
from geopy.geocoders import Nominatim

#initialize the Nominatim object
Nomi_locator = Nominatim(user_agent="My App")

address= "Alexandria,Egypt"

#get the location detail
location = Nomi_locator.geocode(address)
# point_location=Point(lat=location.latitude,lon=location.longitude,alt=location.altitude)
point_location=Point(lat=location.latitude,lon=location.longitude,)
```

```python
# Get daily data for the selected period
data = Daily(point_location, start+timedelta(-1), end)
data = data.fetch()
```

```python
# Plot line chart including average, minimum and maximum temperature
data.plot(y = ['tavg', 'tmin', 'tmax'], kind = 'line')

# Show chart
plt.show()
```
